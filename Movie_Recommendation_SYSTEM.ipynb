```python
# prompt: import numpy and pandas and execute the cell

import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv("/content/drive/MyDrive/Recommendation_complete_tutorial-master (1)/TMDB 5000 movie dataset/tmdb_5000_credits.csv")
```

```python
movies = pd.read_csv("/content/drive/MyDrive/Recommendation_complete_tutorial-master (1)/TMDB 5000 movie dataset/tmdb_5000_movies.csv")
```

```python
movies.rename(index = str, columns = {"id":"movie_id"},inplace =True)
```

```python
movies_credits_merge = movies.merge(credits,on = "movie_id")
movies_credits_merge.head()
```

```python
movies_credits_merge.columns
```

```python
movies_credits_merge.drop(['homepage', 'title_x', 'title_y', 'status', 'production_countries'], axis=1, inplace=True)
```

```python
movies_credits_merge.info()
```

## weighted average  

```python
v =movies_credits_merge['vote_count'] # no.of votes for the movie
R =movies_credits_merge['vote_average'] # average for the movie 0 to 10

C =R.mean()
m = v.quantile(0.70)
m
```

```python
movies_credits_merge["weighted_average"] = ((R*v)+(C*m))/(v+m)
movies_credits_merge.head()
```

```python
movies_sorted_ranking = movies_credits_merge.sort_values("weighted_average",ascending = False)
```

```python
movies_sorted_ranking[['original_title',"vote_count",'vote_average',"weighted_average","popularity"]].head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=movies_sorted_ranking['weighted_average'].head(10),
                    y= movies_sorted_ranking['original_title'].head(10))

# the above two lines will give the graph to us
# no we will do some overaction
plt.xlim(4,10)   # minimum weight values 4 and max 10 to show on plot
plt.title("best moives by weighted average",weight = "bold")
plt.xlabel("weighted average score")
plt.ylabel("Movie title")
plt.savefig('best_movies')
```

```python
# the above code is based on the WEIGHTED AVERAGE ,
# if we consider the popularity also and give 50% weightage

movies_sorted_ranking = movies_credits_merge.sort_values("popularity",ascending = False)

plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=movies_sorted_ranking['popularity'].head(10),
                    y= movies_sorted_ranking['original_title'].head(10))

# the above two lines will give the graph to us
# no we will do some overaction

plt.title("best moives by weighted average",weight = "bold")
plt.xlabel("popularity")
plt.ylabel("Movie title")
plt.savefig('best_movies2')
```

minions is popular from above data


```python
# so now give equal importance to weighted avg and popularity

from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scaled = scaling.fit_transform(movies_credits_merge[['weighted_average','popularity']])
movie_scaled # it just gives scaled values , so created a dataframe with movie scaled data

movie_normalized_df = pd.DataFrame(movie_scaled,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_credits_merge.head()
```

```python
movies_credits_merge[['normalized_weight','normalized_popularity']] = movie_normalized_df
```

```python
movies_credits_merge.head()
```

```python
movies_credits_merge['score'] = movies_credits_merge['normalized_weight']*0.50 + movies_credits_merge['normalized_popularity']*0.50
movies_credits_merge.head()
```

```python
# prompt: based on the score column arange above data in descending order

movies_sorted_ranking = movies_credits_merge.sort_values("score",ascending = False)

movies_sorted_ranking
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=movies_sorted_ranking['score'].head(10),
                    y= movies_sorted_ranking['original_title'].head(10))

# the above two lines will give the graph to us
# no we will do some overaction

plt.title("best moives by weighted average",weight = "bold")
plt.xlabel("score")
plt.ylabel("Movie title")
plt.savefig('best_movies3')
```
